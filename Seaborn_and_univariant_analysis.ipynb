# Heatmaps

```python
flights_df = sns.load_dataset("flights")
flights_df.head()
```

pivot parameter from pandas-
-first parameter takes - row/index,
-Second paramete takes - column,
-Third takes the data


```python
flight_df = pd.pivot(flights_df,index="month",columns="year",values="passengers")
flight_df.head()
```

lighter the color in colorbar - higher the value,
darker the color in colorbar - lower the value

```python
plt.figure(figsize=(10,6))
sns.heatmap(flight_df)
plt.title("Customer using flight is increasing")
```

```python
plt.figure(figsize=(12,6))
#annotation/values will be shown
sns.heatmap(flight_df,annot=True,fmt="d")
plt.title("Customer using flight is increasing")
```

###linecolor & linewidth

```python
plt.figure(figsize=(12,6))
#annotation/values will be shown
sns.heatmap(flight_df,annot=True,fmt="d",linecolor="k",linewidths="3")
plt.title("Customer using flight is increasing")
```

cmap-to choose the color coordination for the heatmap

```python
plt.figure(figsize=(12,6))
sns.heatmap(flight_df,annot=True,fmt="d",linecolor="k",linewidth="2",cmap="Blues")
plt.title("Customer using flight is increasing")
```

remove cbar

```python
plt.figure(figsize=(10,6))
sns.heatmap(flight_df,cbar=False)
plt.title("Customer using flight is increasing")
```

changing coordinates for color bar
using ax->axis and cbar_ax->color bar axis


```python
#hspace-horizontal space
#height ratio - (size of color map, size of color bar):
grid_kws = {"height_ratios":(.5,.05),"hspace":.4}
#f - figure,ax - axix,cbar_ax - color bar axis:
f,(ax,cbar_ax) = plt.subplots(2,gridspec_kw=grid_kws,figsize=(12,6))
ax=sns.heatmap(flight_df,cbar_kws = {"orientation":"horizontal"},ax=ax,cbar_ax=cbar_ax,annot=True,fmt="d")
plt.title("Customer using flight is increasing" )
```

# Univarient analysis

To analyis data based on one value

Histogram

Represents a distribution by forming bins along the range of the data and show number of observations that fall in each bins

>Distplot is used

```python
import seaborn as sns

iris_df = sns.load_dataset("iris")
iris_df.head()
```

```python
sns.histplot(iris_df,x="petal_length",hue="species",kde=True)
```

```python
sns.histplot(iris_df,x="petal_width",hue="species",kde=True)
```

```python
sns.histplot(iris_df,x="sepal_length",hue="species",kde=True)
```

```python
sns.histplot(iris_df,x="sepal_width",hue="species",kde=True)
```

```python
sns.jointplot(x="petal_length",y="petal_width",data=iris_df,kind="scatter")
```

Conclusion:
Having petal width for univarient analysis is better as the graph shows good separation among the species
